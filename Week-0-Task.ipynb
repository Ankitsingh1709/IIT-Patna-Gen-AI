```python
# 1.  Take two numbers as input and print:
# Their sum, difference, product, division, and remainder.

numb1 = int(input("Enter your first number:"))
numb2 = int(input("Enter your second number:"))


print(f"The sum of two numbers are: {numb1 + numb2}")
print(f"The difference of two numbers are: {numb1 - numb2}")
print(f"The product of two numbers are: {numb1 * numb2}")
print(f"The division of two numbers are: {numb1 / numb2}")
print(f"The remainder of two numbers are: {numb1 // numb2}")
```

```python
def check_age():
    age = int(input("Enter your age:"))
    if age > 18:
        return "True"
    else:
        return "False"
    
Result = check_age()
print(Result)
```

```python
# 3. Ask the user to enter their math and science marks (out of 100). If both are greater than 40, print “Pass”, else “Fail”.
def class_check():
    Science = int(input("Enter your Science Marks:"))
    Math = int(input("Enter your Math Marks:"))
    num1 = Science + Math
    if num1 > 40:
        return "Pass"
    else:
        return "Fail"

Result = class_check()
print(Result)
```

```python
# 5. Challenge: Ask for two numbers and print True if the first number is divisible by the second and greater than it.
num1 = int(input("Enter your first number:"))
num2 = int(input("Enter your second number:"))

if num1 % num2 == 0 and num1 > num2:
    print("True")
else:
    print("False")
```

```python
def chek_divisibility():
    num1 = int(input("Enter your first number:"))
    num2 = int(input("Enter your second number:"))

    if num1 % num2 == 0 and num1 > num2:
     return "True"
    else: 
     return "False"
    
result= chek_divisibility()
print(result)
```

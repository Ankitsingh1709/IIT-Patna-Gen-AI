```python
# 1. Write a program to check if a number is even or odd using if-else.
number = int(input("Enter a number:"))

if number % 2 == 0:
    print("number is even")
else:
    print("number is odd")
    
```

```python
### 2. Loop through numbers 1 to 10 and print only even numbers using continue.

for i in range(1 , 11):
    if i % 2 == 0:
        print(i)
    else:
        continue
```

```python
# 2. Loop through numbers 1 to 10 and print only even numbers using continue.

counter = 1

while counter <= 10:
    if counter % 2 != 0:
     print(f"the number is odd continue the loop {counter}")
     counter += 1
     continue    
    if counter % 2 == 0:
       print(counter) 
    counter += 1 
    # break

print(f"**the loop end at this value of counter {counter}**")  
```

```python
# 3. Ask the user to guess a number between 1–5. Keep asking until the guess is correct (use while and break)


guess_number = 3

while True:
    num = int(input("please guess the number betwwen 1 to 5"))
    if num != guess_number:
     print(f"You gussed the wrong number {num} please try again")
     continue

    else:
     print(f"Congrulation you have gussed the correct number: {num}")
    break
    

    
```

```python
# 4. Use a for loop to calculate the sum of all numbers between 1 and 100.

numbers = []

for i in range(1,101):
    numbers.append(i)
print(numbers)

total = sum(numbers)
print(total)
```

```python
total = 0

for i in range(1, 101):
    total += i

print(total)
```

```python
### Write a program to count how many vowels are in a given string using a for loop.

vowels = ["a", "e", "i", "o", "u"]

statment  = input("Enter a statment:" ).lower()

count = 0

for i in statment:
    if i in vowels:
        count += 1


print("the number of vowels in statment are:", count )
    
```
